# PROJECT : PREDICTING MERGER AND ACQUISITION

##### In this project our goal is to predict if a merger and acquisition will occur.The model is built on the dataset dataMA.csv witch contain 
##### historical data about merger and acquisition in Europe, US, and  Asia.So we are mainly interested in the True-Positive (When the model 
##### predict that there will be a merger and acquisition and it really occure) accuracy of the model rather than it's global accuracy.

### Importing needed libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

```

## Import tools needed for visualization

```python

import graphviz
from sklearn import tree
```

## Cleaning and preparing the dataset

```python

df = pd.read_csv("dataMA.csv",sep=";")

df['y']=np.where(df['Deal Status'] == "Completed", 1, 0)

df['isfriendly']=np.where(df['Nature of Bid'] == "Friendly", 1, 0)

df['isCash']=np.where(df['Payment Type']== "Cash", 1, 0)

df=df[df['isCash']>0]

df=df[df['Percent Sought']>40]

df.dropna(subset = ['isCash','isfriendly','Announced Total Value (mil.)', 'Premium to 1-Month Avg Px (%)','Percent Sought','Percent Owned','Net Debt'], inplace=True)

df['R']=df['Target ISIN'].str[:2]

df['A']=df['Acquirer ISIN'].str[:2]

df['sameRegion']=np.where(df['R'] == df['A'], 1, 0)

df = df[df["Deal Status"] != "Pending" ]

#Feature selection

X=df[['sameRegion','isCash','isfriendly','Announced Total Value (mil.)', 'Premium to 1-Month Avg Px (%)','Percent Sought','Percent Owned','Net Debt']] # Features

y=df['y']

#Type conversion
X.astype(np.float32)

feature_list = list(X.columns)

```

## Split, Fit , and Predict:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test

#Importing the random forest that we will use for the modelisation 
from sklearn.ensemble import RandomForestClassifier


#Handing our umbalanced dataset and fitting the randomforest with the train data

clf=RandomForestClassifier(n_estimators=1000,class_weight={0:1,1:0.1},min_samples_leaf=100,max_depth=15)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)

y_pred_proba=clf.predict_proba(X_test)

#To improve the accuracy will consider only the "true-positives" where the accuracy of the model is higher than 0.6 
for i in range(len(y_pred_proba)):
    if y_pred_proba[i][1]<=0.6:
        y_pred[i]=0

#Import scikit-learn metrics module for accuracy calculation

from sklearn import metrics

# Model Accuracy, how often is the classifier correct?

#print("Accuracy:",metrics.accuracy_score(y_test, y_pred))


cm=(confusion_matrix(y_test, y_pred))

plt.figure(figsize = (3,3))

sns.heatmap(cm, annot=True)

plt.show()


#print(cm[0][0]/(cm[1][0]+cm[0][0]))

print("Accuracy when the model predict that trere will be a Merger/Acquisition : ", round(1-(cm[0][1]/(cm[0][1]+cm[1][1])),2))

#print((cm[0][1]+cm[0][0])/(y_test.size))




```

## Feature Importances:

```python
# Get numerical feature importances

importances = list(clf.feature_importances_)

# List of tuples with variable and importance

feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]

# Sort the feature importances by most important first

feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances

[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];


import matplotlib.pyplot as plt

# Set the style

plt.style.use('fivethirtyeight')

# list of x locations 
import matplotlib.pyplot as plt
# Set the style
plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(importances)))
# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')
# Tick labels for x axis
plt.xticks(x_values, feature_list, rotation='vertical')
# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
plt.show()
```
